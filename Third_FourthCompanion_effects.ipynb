```python
import numpy as np
import matplotlib.pyplot as plt
import os

import astropy.units as u
import astropy.constants as const
import mesaPlot as mp

home_dir = os.path.expanduser("~")
data_root_dir    = home_dir + '/ceph/ksharpe/results/' #'/ceph/MESAoutput/ksharpe_proj/'
inlist_dirs = ['dense_mass_hastings_grid', 'dens_mass_grid_run' 'dense_mass_div3winds_grid',  'CHE_x3winds']


def Kepler(m1 = 1*u.Msun, m2 = 1*u.Msun, P = 1*u.day):
    m1 = m1.to(u.kg)
    m2 = m2.to(u.kg)
    P = P.to(u.s)
    a = (const.G * (m1 + m2) * (P**2/4*np.pi**2) )**(1/3)
    return a.to(u.Rsun)
```

# equation 32 from Hamers and Lai 2017 https://arxiv.org/abs/1705.02334


```python
# equation 32 from Hamers and Lai 2017 https://arxiv.org/abs/1705.02334
def beta_22(i_init = 90, a1 =1*u.Rsun, a2 = 1*u.Rsun, 
            m0 = 1*u.Msun, m1 = 1*u.Msun, m2 = 1*u.Msun, m3 = 1*u.Msun):
    return (3./4) * np.cos(np.radians(i_init)) * ((a2/a1)**(3./2)) * ((m0 + m1) / (m2 + m3))**(3/2)
```

# For current day parameters
20 day orbit, 60 Msun stars

```python
mA, mB = 61*u.Msun, 66*u.Msun
mC = 30*u.Msun
mD = 3*u.Msun # or 40 *u.Msun
P_AB = 19.2*u.day
P_CD = 96.5*u.day

a_AB = Kepler(m1 = mA, m2 = mB, P = P_AB)
a_CD = Kepler(m1 = mC, m2 = 66*u.Msun, P = P_CD)

inclinations = np.linspace(0,90)
beta_lowMd = beta_22(i_init = inclinations, a1 =a_AB, a2 = a_CD, 
            m0 = mA, m1 = mB, m2 = mC, m3 = 3*u.Msun)

plt.plot(inclinations, beta_lowMd, label = 'MD = 3Msun')

beta_highMd = beta_22(i_init = inclinations, a1 =a_AB, a2 = a_CD, 
            m0 = mA, m1 = mB, m2 = mC, m3 = 40*u.Msun)

plt.plot(inclinations, beta_highMd, label = 'MD = 40Msun')

plt.axvline(86, c = 'k', linestyle = 'dashed')

plt.legend()
plt.xlabel('inclination (degrees)') 
plt.ylabel(r'$\beta_{2+2}$')
plt.show()


print('for MD = 3Msun, beta=', beta_22(i_init = 86, a1 =a_AB, a2 = a_CD, 
            m0 = mA, m1 = mB, m2 = mC, m3 = 3*u.Msun) )
print('for MD = 40Msun, beta=', beta_22(i_init = 86, a1 =a_AB, a2 = a_CD, 
            m0 = mA, m1 = mB, m2 = mC, m3 = 40*u.Msun) )
```

# The relativistic apsidal motion timescale 

```python
# relativistic precession time-scale (eq 25 from Hamers and Lai 2017)

def calculate_tPN(Porb_1, e, a, m0, m1):
    # Calculate rg
    Grav_const = const.G.to(u.Rsun**3 / (u.Msun * u.yr**2))
    light_speed = const.c.to(u.Rsun / u.yr)
    rg = Grav_const * (m0 + m1) / (light_speed**2)

    # Calculate t
    t = 1/3 * Porb_1 * (1 - e**2) * a / rg

    return t.to(u.yr)


## Assuming current day masses and orbits
P_AB = 20*u.day
e = 0.27
MA = 61*u.Msun
MB = 66*u.Msun
a_AB = Kepler(m1 = mA, m2 = mB, P = P_AB)

# Use the function
t = calculate_tPN(P_AB, e, a_AB,MA,MB)
print('relativistic apsidal motion timescale current day:', t)



## Assuming future evol with fiducial model
P_AB = 55*u.day
e = 0.27
MA = 37*u.Msun
MB = 37*u.Msun
a_AB = Kepler(m1 = mA, m2 = mB, P = P_AB)

# Use the function
t = calculate_tPN(P_AB, e, a_AB,MA,MB)
print('relativistic apsidal motion timescale Future Fiducial:', t)



## Assuming future evol with enhanced wind model
P_AB = 220*u.day
e = 0.27
MA = 19*u.Msun
MB = 19*u.Msun
a_AB = Kepler(m1 = mA, m2 = mB, P = P_AB)

# Use the function
t = calculate_tPN(P_AB, e, a_AB,MA,MB)
print('relativistic apsidal motion timescale Future enhanced wind:', t)
```
