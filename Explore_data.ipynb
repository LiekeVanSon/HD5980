```python

def read_file(hist_profile, basedir, pnum=10):
    """
    hist_profile ---> string {history, profile}
    basedir      ---> string {location of file}
    pnum         ---> int    {profile number}
    """
    # Check if profile number is provided
    if hist_profile == 'profile':
        hist_profile += str(pnum)

    # Open file and get column names
    with open(basedir + hist_profile + '.data', "r") as file:
        run_info_keys, run_info_vals, ColName = None, None, None
        star_age = -999
        for i, line in enumerate(file):
            # print(100*'*', line, '\n')
            if i == 1:
                run_info_keys = np.array(line.split())
            elif i == 2:
                run_info_vals = np.array(line.split())
                if 'profile' in hist_profile:
                    star_age = np.float32(run_info_vals[4]) # Convert seconds to years
            elif i == 5:
                ColName = np.array(line.split())
                break

    # Create a dictionary of run information
    run_info_dict = dict(zip(run_info_keys, run_info_vals))
            
    # Load data into a table
    try:
        data = np.loadtxt(basedir + hist_profile + '.data', comments='#', converters=None, skiprows=6, unpack=True)
        #print('shape of data', np.shape(data))
    except:
        print('Could not read file:', basedir + hist_profile + '.data')
        return None, None, None
        
    t = Table()
    for j, nome in enumerate(ColName):
        try:
            t[nome] = data[j]
        except:
            print('You couldnt read this file:', basedir +hist_profile+'.data' )
            return None,None
        
    return t, run_info_dict, star_age



## Folders to choose from
folder_list = ['dense_mass_div3winds_grid', 'dense_mass_grid_run', 'dense_mass_hastings_grid',\
               'dense_mass_standard_grid', 'dense_mass_x3winds_grid', 'mesa_models']


folder = 'dense_mass_x3winds_grid' #'dense_mass_hastings_grid'
# Print available runs in this folder:
# print(os.listdir(root_dir +  folder + '/'))
    
Nsim = 290

nlist = []
for N in range(0,500):
#     print('N', N)
    file_dir    = root_dir +  folder + '/' +str(N) + '/LOGS1/' 
    
    try:
        table, run_info_dict, star_age  = read_file('history', file_dir)

        # display(table)
        # table.info()
#         print('MASS = ', table['star_mass'][0], '\n')
#         print(run_info_dict)
        if table['star_mass'][0] == 188.364908949 :#79.4328234724: #100.:
            print('N', N, 'MASS = ', table['star_mass'][0], '\n')
            nlist.append(N)

    except:
        continue

print(nlist)

```
